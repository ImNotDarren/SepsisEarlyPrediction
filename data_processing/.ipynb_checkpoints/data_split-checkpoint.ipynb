```python
import os
import pandas as pd
import numpy as np
import random
import pickle
```

```python
data_path = '../dataset/training_setA/'
train_path = '../dataset/raw/train/'
val_path = '../dataset/raw/val/'
test_path = '../dataset/raw/test/'

pid = sorted(os.listdir(data_path))[1:]

train_num = round(0.7*len(pid)) + 1
val_num = round(0.15*len(pid))
test_num = round(0.15*len(pid))
```

```python
len(pid) - train_num - val_num - test_num
```

```python
# data split
train = random.sample(pid, train_num)
val = random.sample(set(pid)-set(train), val_num)
test = set(pid) - set(train) - set(val)

for p in train:
    df = pd.read_csv(data_path + p, sep='|')
    df.to_csv(train_path + p, sep='|', index=False)

for p in val:
    df = pd.read_csv(data_path + p, sep='|')
    df.to_csv(val_path + p, sep='|', index=False)
    
for p in test:
    df = pd.read_csv(data_path + p, sep='|')
    df.to_csv(test_path + p, sep='|', index=False)
```

```python
train_concat = pd.DataFrame()

# train_concat
patient_id = 1
for p in train:
    df = pd.read_csv(train_path + p, sep='|')
    rows = df.shape[0]
    df['patient_id'] = pd.Series([patient_id for x in range(rows)])
    train_concat = pd.concat([train_concat, df])
    patient_id += 1
column = train_concat.columns.tolist()[0]
train_concat = train_concat.drop(columns=[column])
```

```python
train_concat
```

```python
train_concat.to_csv('../dataset/train_concat.csv', sep='|', index=False)
```

```python
val_concat = pd.DataFrame()

# val_concat
patient_id = 1
for p in val:
    df = pd.read_csv(val_path + p, sep='|')
    rows = df.shape[0]
    df['patient_id'] = pd.Series([patient_id for x in range(rows)])
    val_concat = pd.concat([val_concat, df])
    patient_id += 1
```

```python
val_concat
```

```python
val_concat.to_csv('../dataset/val_concat.csv', sep='|', index=False)
```

```python
test_concat = pd.DataFrame()

# test_concat
patient_id = 1
for p in test:
    df = pd.read_csv(test_path + p, sep='|')
    rows = df.shape[0]
    df['patient_id'] = pd.Series([patient_id for x in range(rows)])
    test_concat = pd.concat([test_concat, df])
    patient_id += 1
```

```python
test_concat.to_csv('../dataset/test_concat.csv', sep='|', index=False)
```
